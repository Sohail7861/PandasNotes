21.Find Unique Values From Genre 

```python
datas['Genre']
```

```python
list1=[]
for value in datas['Genre']:
    list1.append(value.split(','))
list1
```

```python
unique=[]
for i in list1:
    for j in i:
        unique.append(j)
unique
```

```python
final=[]
for i in unique:
    if i not in final:
        final.append(i)
final
```

```python
len(final)
```
